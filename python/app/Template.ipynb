  function storeNotebookName() {
    var name = IPython.notebook.get_notebook_name();
      if(name.search(/template/i)!=-1) {
        if(name.search(/copy/i)!=-1) {
            document.getElementById("MessageBar").innerHTML = "<font color='red'>Please rename the notebook.  </font>";
            while(1==1) {
               var newName=prompt("Please enter a new name for the notebook.\n\nDo not use spaces, or any special character. You can use _-+ if you like.\n\n For example, GRB_080916C_100MeV-10GeV_FRONT+BACK is a valid name.","");
                if(validate(newName)) {
                  break;
                      } else { continue };
            }
        } else {
            document.getElementById("MessageBar").innerHTML = "<font color='red'>This is the template, making a copy...</font>";
            makeCopy();
        }
        throw "Notebook is template";
    }
    //Store the name in the kernel and set the new pfiles directory
    IPython.notebook.kernel.execute("notebookName = '"+name+"'");
    IPython.notebook.kernel.execute("import os,subprocess");
    IPython.notebook.kernel.execute("os.environ['NOTEBOOK_NAME'] = notebookName");
    IPython.notebook.kernel.execute("newPfiles                   = '%s/Notebooks/%s_pfiles' %(os.environ['BASEDIR'],notebookName)");
    IPython.notebook.kernel.execute("try:\n    os.makedirs(newPfiles)\nexcept:\n    pass");
    IPython.notebook.kernel.execute("os.environ['PFILES'] = '%s;%s' % (newPfiles,os.environ['PFILES'].split(';')[1])");
    IPython.notebook.kernel.execute("os.environ['OUTDIR'] = '%s/Notebooks/%s_outdir' %(os.environ['BASEDIR'],notebookName) ");
    IPython.notebook.kernel.execute("try:\n    os.makedirs(os.environ['OUTDIR'])\nexcept:\n    pass");
    //Import the utilities
    IPython.notebook.kernel.execute("from gtgrb import *");
    IPython.notebook.kernel.execute("os.environ['GUIPARAMETERS']='yes'");
    IPython.notebook.kernel.execute("from GTGRB.Redirector import *");
    IPython.notebook.kernel.execute("start_redirect()");
    IPython.notebook.select_next();
  }
    function gracefulDead(timerId,n) {
        //alert("gracefulDead");
        n = n+1;
        try {
            storeNotebookName();
            window.clearInterval(timerId);
            document.getElementById("MessageBar").innerHTML = "You are now ready to start GtGRB.";
        } catch(err) { }
    }
    
    function checkTitleChange(titleTimer) {
      var name = IPython.notebook.get_notebook_name();
      if(name.search(/template/i)==-1) {
          //Title has changed
          window.clearInterval(titleTimer);
          var n=999;
          gracefulDead(null,n);
      }
    }
    function validate(name) {
        if(name==null) {
          return false;
        }
	        re=/[^a-zA-Z0-9_+-]+/;

	        if(re.test(name)) {
	            alert("Invalid Entry:\nOnly Alphabetic characters, minus, plus or underscore allowed!Please correct the name.");
	            return false;
	        } else {
                //Good name!
                changeName(name);
                return true;
            }
	}
    function changeName(name) {
      IPython.notebook.set_notebook_name(name);
      IPython.notebook.save_notebook();
    }
    function makeCopy() {
            var notebook_id = IPython.notebook.get_notebook_id();
            var url = $('body').data('baseProjectUrl') + notebook_id + '/copy';
            window.open(url,'_self');
        }
    function handleTimer() {
        //alert(timerId);
      var n=0;
      gracefulDead(timerId,n);
    }
    var timerId = setInterval(handleTimer,500);
</script>
